```python
import string

query = "Buy mobile phone buy phone online"
query = query.lower()
query = query.translate(str.maketrans('', '', string.punctuation))
words = query.split()

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

result = {}
for word, count in freq.items():
    if count > 1:
        result[word] = count

print(result)
```
