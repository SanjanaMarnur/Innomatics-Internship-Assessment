```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
```

```python
my_dict = {}
for email in emails:
  domain_name = email.split("@")[1]
  if my_dict.get(domain_name) == None:
    my_dict[domain_name] = 1
  else:
    my_dict[domain_name] += 1


total = len(emails)
for domain_name,freq in my_dict.items():
  percentage = f"{int((freq/total)*100)}%"
  my_dict[domain_name] = percentage
  print(f"{domain_name} : {percentage}")
```
