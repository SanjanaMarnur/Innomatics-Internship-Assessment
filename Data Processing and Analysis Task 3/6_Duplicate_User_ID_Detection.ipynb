```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
```

```python
my_dict = {}
for i, user in enumerate(user_ids):
  if(my_dict.get(user) == None):
    my_dict[user] = 1
  else:
    my_dict[user] += 1

result = [f"{user} → {freq} times" for user, freq in my_dict.items() if freq > 1]
for res in result:
  print(res)
```
