```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
```

```python
my_dict = {}
for i, value in enumerate(sensor_readings):
  if value % 2 == 0 :
    my_dict[i] = value

print(my_dict)
```

```python
list = [(hour, value) for hour, value in my_dict.items()]
print("Valid Sensor Readings (Hour, Value):")
print(list)
```
